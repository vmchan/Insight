```python
# Load data
df_all = pd.read_csv('IDS_SET2.txt', sep='|', na_values='NULL', encoding='utf-8-sig') # need to specify encoding to get rid of BOM at start of text file
df_all.head()
```

```python
# Check for empty rows 
original_shape = df_all.shape[0]
print('There are {} rows for all orgs.'.format(original_shape)) # Check size of dataframe before cleaning
df_all.dropna(how='all', inplace = True) # Clean data by removing rows where everything in null. Just the last row qualifies
print('There were {} empty row(s) for all orgs.'.format(original_shape-df_all.shape[0]))
```

```python
# Remove row with UK zip. It can be found with df_all[df_all.values == 'CB2 4QR']
df_all = df_all[df_all.ZIP != 'CB2 4QR']
```

```python
# Convert check box 'Y' strings and NaNs to bools so I can turn them into a numpy array
df_all['TERMINATED'].replace('Y', True, inplace=True)
df_all['TERMINATED'].fillna(False, inplace=True)

# Horribly, I can't find a good way to do this for multiple columns, but not the whole dataframe, so repeat as needed:
df_all['FINAL_RET_CBX'].replace('Y', True, inplace=True)
df_all['FINAL_RET_CBX'].fillna(False, inplace=True)

df_all['REPORT_COSTS_CBX'].replace('Y', True, inplace=True)
df_all['REPORT_COSTS_CBX'].fillna(False, inplace=True)

# Make NaNs sci-kit learn friendly 
df_all.fillna(0, inplace=True)

# Clean up the zipcode data. 
df_all['ZIP'] = df_all['ZIP'].map(lambda x: str(x)[:5])

df_all.head()
```

```python
# Try to classify everything using logistic regression

cv = StratifiedKFold(y, n_folds=10)
#classifier = ensemble.RandomForestClassifier()
classifier = linear_model.LogisticRegression()

mean_tpr = 0.0
mean_fpr = np.linspace(0, 1, 100)
all_tpr = []

plt.figure(figsize=(16, 16))

for i, (train, test) in enumerate(cv):
    probas_ = classifier.fit(X[train], y[train]).predict_proba(X[test])
    # Compute ROC curve and area the curve
    fpr, tpr, thresholds = metrics.roc_curve(y[test], probas_[:, 1])
    mean_tpr += interp(mean_fpr, fpr, tpr)
    mean_tpr[0] = 0.0
    roc_auc = metrics.auc(fpr, tpr)
    plt.plot(fpr, tpr, lw=1, label='ROC fold %d (area = %0.2f)' % (i, roc_auc))

plt.plot([0, 1], [0, 1], '--', color=(0.6, 0.6, 0.6), label='Luck')

mean_tpr /= len(cv)
mean_tpr[-1] = 1.0
mean_auc = metrics.auc(mean_fpr, mean_tpr)
plt.plot(mean_fpr, mean_tpr, 'k--',
         label='Mean ROC (area = %0.2f)' % mean_auc, lw=2)

plt.xlim([-0.05, 1.05])
plt.xticks(fontsize=16)
plt.ylim([-0.05, 1.05])
plt.yticks(fontsize=16)
plt.xlabel('False Positive Rate', fontsize=16)
plt.ylabel('True Positive Rate', fontsize=16)
plt.title('Cross validation using stratified k-folds', fontsize=16)
plt.legend(loc="lower right", fontsize=16)
plt.savefig('LogisticRegression.png')
```

```python
# Break data into test and training sets using stratified K folds since it is so biased to being alive
from sklearn import cross_validation
def stratified_cv(X, y, clf_class, shuffle=True, n_folds=10, **kwargs):
    stratified_k_fold = cross_validation.StratifiedKFold(y, n_folds=n_folds, shuffle=shuffle)
    y_pred = y.copy()
    for i, j in stratified_k_fold:
        X_train, X_test = X[i], X[j]
        y_train = y[i]
        clf = clf_class(**kwargs)
        clf.fit(X_train,y_train)
        y_pred[j] = clf.predict(X_test)
    return y_pred
```

```python
random_forest_conf_matrix = metrics.confusion_matrix(y, stratified_cv(X, y, ensemble.RandomForestClassifier))
logistic_reg_conf_matrix = metrics.confusion_matrix(y, stratified_cv(X, y, linear_model.LogisticRegression))
dumb_conf_matrix = metrics.confusion_matrix(y, [0 for ii in y.tolist()]); # ignore the warning as they are all 0

conf_matrix = {
                1: {
                    'matrix': random_forest_conf_matrix,
                    'title': 'Random Forest',
                   },
                2: {
                    'matrix': logistic_reg_conf_matrix,
                    'title': 'Logistic Regression',
                   },
                3: {
                    'matrix': dumb_conf_matrix,
                    'title': 'Dumb',
                   },
}
```

```python
from lifelines.plotting import plot_lifetimes

fig = plt.figure(figsize=(16, 16)) 
plt.xlabel('Year', size=16)
plt.ylabel('NPO ID', size=16)
plt.xticks(range(first_year, last_year, 1), fontsize =16)
plt.yticks(range(50), npoids[:50], fontsize =16)
plot_lifetimes(durations[:50], birthtimes=birthyears[:50], event_observed=observed[:50])
fig.savefig('lifelines.png', bbox_inches='tight')
```
